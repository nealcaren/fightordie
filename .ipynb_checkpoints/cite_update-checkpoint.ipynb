```python
from glob import glob
import re
```

```python


def clean(fn):
    
    # Figure out if it already has a front matter:
    
    
    
    meta_d = {'journal': 'The Crisis'}
    
    with open(fn,'r') as infile:
        md = infile.read()
        
    if 'pages:' in md and 'volume:' in md:
        md = md.split('-->\n')[-1]
    print(md)  
    title_line = re.findall('^\#(.*?) \(', md)[0]
    title = title_line.strip().replace('"', "'")
    
    cite  = re.findall('\*Citation.*?$', md)[0]
    while '  ' in cite:
            cite = cite.replace('  ', ' ')
    cite = cite.replace('Dubois', 'Du Bois')
    
    if 'Du Bois, W.E.B.' in cite:
        meta_d['author'] = 'Du Bois, W.E.B.'
    elif 'Dr. Du Bois Resigns' in cite:
        meta_d['author'] = 'National Association for the Advancement of Colored People'
    else:
        meta_d['author'] = cite.split('*')[2].split('.')[0].strip()
    meta_d['year'] = cite.split('. "')[0].split('"')[0].replace('.','').strip()[-4:]
    meta_d['title'] = title
    
    t = cite.split(' ')
    if '(' in t[-1]:
        placement = t[-1]
    else:
        placement = t[-2] + t[-1]
    placement = placement.replace('.','')
    pages = placement.split(':')[-1]
    meta_d['pages'] = pages.replace(',', ', ')
    meta_d['volume'] =placement.split('(')[0]
    meta_d['issue'] = placement.split('(')[1].split(')')[0]
   
    title_text = f'''# {meta_d["title"]} ({meta_d["year"]})'''
    new_cite_line = f'''*Citation:* {meta_d["author"]}. {meta_d["year"]}. "{meta_d["title"]}." *{meta_d["journal"]}*. {meta_d["volume"]}({meta_d["issue"]}):{meta_d["pages"]}.'''
    new_cite_line = new_cite_line.replace('.. 1', '. 1')
    
    for row in md.splitlines():
        if row[:2] == '# ':
            old_title_line = row
            
    #update cite
    
    md = md.replace(old_title_line, title_text)
    
    c = f'''<!--
title:   {meta_d["title"]}
author:  {meta_d["author"]}
journal: {meta_d["journal"]}
year:    {meta_d["year"]}
volume:  {meta_d["volume"]}
issue:   {meta_d["issue"]}
pages:   {meta_d["pages"]}
-->
'''
    # add frontmatter
    md = c + md
    
    with open(fn,'w') as outfile:
        outfile.write(md)
```

```python
fns = glob('Volumes/**/*.md', recursive=True)
fns = sorted(fns)

for fn in fns:
    print(fn)
    clean(fn)
```
