```python
# Clean up page titles



def clean(fn):
    with open(fn,'r') as infile:
        html = infile.read()

    ugh = '''&#8212; Dare You Fight&amp;#58;&lt;br/&gt;W.E.B. Du Bois&lt;br/&gt;in The Crisis<'''
    ugh2= ''':\n   <br/>\n   W.E.B. Du Bois\n   <br/>\n   in The Crisis\n  <'''

    html = html.replace(ugh,'<')
    html = html.replace(ugh2,'<')
    
    #toc fix
    
    for href in ['index.html', '#', '../index.html', '../../index.html', '../../../index.html']:
        

        link = f'''<a class="reference internal" href="{href}">
                    Dare You Fight: W.E.B. Du Bois in The Crisis
                </a>'''
        
        new_link = link.replace(': W.E.B. Du Bois in The Crisis','')
        
        html = html.replace(link, new_link)
        
    
    # Remove weird symbol on hover.
    html = html.replace('>¶<', '><')
    html = html.replace('>#<', '><')
    
    with open(fn,'w') as outfile:
        outfile.write(html)

    
    
```

```python

fns = glob('_build/html/**/*.html', recursive=True)


for fn in fns:
    clean(fn)
```
